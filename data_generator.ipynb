```python
import pandas as pd;
from tqdm import tqdm;
import numpy as np;
import random

def data_generator (nb_data = 10000 , 
                    Fields_cho = [0, 1, 2], # Business 0 / Med 1 / Other 2
                    local_cloud = [0, 1],
                    low_medium_high = [0, 1, 2],
                    vul_id_min = 0,
                    vul_id_max = 10,
                    cybersecu_investment = 0.14,
                    number_employees = (0, 1, 2, 3),
                    min_employee = 10,
                    max_employee = 5000,
                    revenue_min = 1,
                    revenue_max = 1000):
    attributes = ["Fields",  "Data Storage", "Access Control", "Vulnerability Identification", "IT Support", "Cybersecurity Investment", "Cybersecurity Awareness", "Number of employees", "Revenue"]
    df = pd.DataFrame (columns = attributes)
    
    for i in tqdm (range (0 , nb_data )) :

        Fields = np.random.randint(0, 2)
        
        datastorage = np.random.choice(local_cloud)
                    
        #Employee option 1 : use categories
        #93% of enterprises are micro sme -> 0.93
        #1% are large -> 0.01
        #split up the small and medium -> 0.03
        nb_employee_1 = np.random.choice (np.arange(0, 4), p=[0.93, 0.03, 0.03, 0.01])
        
        #Employee option 2 : use the numbers. 500 seems too low
        nb_employee = np.random.randint(0, 5000)
        
        #In percent : 13.7 would be the ideal one, with a standard deviation of 6 (for now, to justify)
        cybersecuinvest = np.random.normal(10, 6)
        if cybersecuinvest < 0: cybersecuinvest = 0
        
        #Access control not conditional 
        accesscontrol = np.random.normal(1.5,1)
        accesscontrol = int(accesscontrol)

        #ToRefine, not sure about the numbers yet
        vulnerabilityid = np.random.randint(vul_id_min, vul_id_max)
        
        #IT Support, conditional on employees
        if nb_employee <= 100 :
            it_support = 0
        elif 100 < nb_employee <=500 and random.random()<=0.8:
            it_support = np.random.choice([0, 1])
        elif 500 < nb_employee <=2500 and random.random()<=0.8:
            it_support = np.random.choice([0, 1, 2])
        else:
            it_support = np.random.choice([1, 2])
      
        #Cybersecurity awareness conditional on ITSupport (so on employees)
        if it_support == 0:
            cyber_awareness = 0
        elif it_support == 1 :
            cyber_awareness = np.random.choice([0, 1])
        else :
            cyber_awareness = np.random.choice([1, 2])
        
        #Revenue: ToRefine (in millions for now) conditionnal on employees
        revenue = (0.15 + np.random.normal(0, 0.05)) * nb_employee 
        if revenue <= 10 : 
          if cybersecuinvest >= 5 : cybersecuinvest = 5

        #Risk : to calculate / or not. for now 0
    
        df.loc[i] = [Fields,  datastorage, accesscontrol, vulnerabilityid, it_support, cybersecuinvest, cyber_awareness, nb_employee, revenue ]
        

    df.to_csv('out.csv',index=False)

    return df
    


data_generator()
```
